```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def normal(x):
    return (np.exp(-x*x/2)/(np.sqrt(2*np.pi))) # Gaussian distribution with zero mean and variance 1
#can send variance and mean as arguments as well
```

```python
n = 100000
y = np.arange(-3,3,0.01)
x = np.zeros(n)
x[0] = 3
kount = 0
```

```python
for i in range(1,n):
    xstar = x[i-1] + np.random.normal()
    alpha = np.random.uniform()
    ratio = normal(xstar)/normal(x[i-1])
    if alpha<min(1,ratio):               # acceptance step
        x[i]=xstar
        kount = kount+1
    else:
         x[i]=x[i-1]

print('Ratio of samples accepted to the total samples')       
print(kount/n)
```

```python
plt.hist(x,density=True,label='MCMC Samples')
plt.plot(y,normal(y),label='actual dist') 
plt.title('Histogram of values of posterior by Metropolis Hastings algorithm')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
```
