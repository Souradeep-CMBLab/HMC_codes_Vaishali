```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epsi = 0.3
```

```python
def pe(x):
    return x
def ke(p):
    return p
```

```python
n = 20
x1 = np.zeros(n)
p1 = np.zeros(n)
x2 = np.zeros(n)
p2 = np.zeros(n)
x3 = np.zeros(n)
p3 = np.zeros(n)
x4 = np.zeros(n)
p4 = np.zeros(n)
x1[0]=0
p1[0]=1
x2[0]=0
p2[0]=1
x3[0]=0
p3[0]=1
x4[0]=0
p4[0]=1
```

```python
#Euler method
for i in range(0,n-1):
    p4[i+1] = p4[i] - epsi*pe(x4[i])
    x4[i+1] = x4[i] + epsi*ke(p4[i])    
   
```

```python
#Modified Euler method
for i in range(0,n-1):
    p1[i+1] = p1[i] - epsi*pe(x1[i])
    x1[i+1] = x1[i] + epsi*ke(p1[i+1])
```

```python
#Leapfrog method stepsize = 0.3
for i in range(0,n-1):
    temp = p2[i] - epsi*pe(x2[i])/2
    x2[i+1] = x2[i] + epsi*ke(temp)
    p2[i+1] = temp - epsi*pe(x2[i+1])/2
```

```python
#Leapfrog method stepsize = 1.2
epsi = 1.2
for i in range(0,n-1):
    temp = p3[i] - epsi*pe(x3[i])/2
    x3[i+1] = x3[i] + epsi*ke(temp)
    p3[i+1] = temp - epsi*pe(x3[i+1])/2
```

```python
figure, axis = plt.subplots(2, 2)
Drawing_uncolored_circle = plt.Circle( (0, 0 ),1 ,fill = False )
axis[0,0].plot(x4,p4)
axis[0,0].add_artist( Drawing_uncolored_circle )
axis[0,0].set_title("Euler's Method, stepsize = 0.3")
axis[0,0].set_xlabel("position(x)")
axis[0,0].set_ylabel("momentum(p)")
Drawing_uncolored_circle = plt.Circle( (0, 0 ),1 ,fill = False )
axis[0,1].plot(x1,p1)
axis[0,1].add_artist( Drawing_uncolored_circle )
axis[0,1].set_title("Modified Euler's Method")
axis[0,1].set_xlabel("position(x)")
axis[0,1].set_ylabel("momentum(p)")
Drawing_uncolored_circle = plt.Circle( (0, 0 ),1 ,fill = False )
axis[1,0].plot(x2,p2)
axis[1,0].add_artist( Drawing_uncolored_circle )
axis[1,0].set_title("Leap frog Method, stepsize = 0.3")
axis[1,0].set_xlabel("position(x)")
axis[1,0].set_ylabel("momentum(p)")
Drawing_uncolored_circle = plt.Circle( (0, 0 ),1 ,fill = False )
axis[1,1].plot(x3,p3)
axis[1,1].add_artist( Drawing_uncolored_circle )
axis[1,1].set_title("Leap frog Method, stepsize = 1.2")
axis[1,1].set_xlabel("position(x)")
axis[1,1].set_ylabel("momentum(p)")
figure.tight_layout()
```
