```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def hamil(x,p):
    return np.exp(-(x*x/2 + p*p/2))  #1-D Harmonic oscillator
```

```python
def pe(x):
    return x

def ke(x):
    return x
```

```python
epsi = 0.2
n=100000
x = np.zeros(n)
p = np.zeros(n)
x[0] = 1
kount = 0
```

```python
for i in range(0,n-1):
    p[i] = np.random.normal()
    temp = p[i]-epsi*pe(x[i])/2                        
    xtemp = x[i]+epsi*ke(temp)            # Leapfrog to calculate the next sample in the chain
    ptemp = temp-epsi*pe(xtemp)/2
    ratio = hamil(xtemp,ptemp)/hamil(x[i],p[i])    
    alpha = np.random.uniform()
    if alpha<min(1,ratio):   # acceptance step
        x[i+1]=xtemp
        kount = kount + 1
    else:
        x[i+1]=x[i]
print('Ratio of samples accepted to the total number of samples') 
print(kount/n)
```

```python
figure, axis = plt.subplots()
#Drawing_uncolored_circle = plt.Circle( (0, 0 ),3.5 ,fill = False )
#axis.add_artist( Drawing_uncolored_circle )
axis.plot(x,p,'o')

#plt.title('Histogram of x values from HMC')
plt.xlabel('x')
plt.ylabel('momentum')
plt.title('Phase space of a Harmonic Oscillator using HMC ')
```
